```python
# 3-1 필수과제
import pandas as pd                                         # 판다스 불러오기
import numpy as np                                          # 넘파이 불러오기
df_excel = pd.read_excel('관서별 5대범죄 발생 및 검거.xlsx')  # 엑셀 파일 불러오기
print(df_excel)                                             # 엑셀 파일 출력
```

```python
# 3-2 필수과제
ps_list =  [(df_excel['관서명'] == '서대문서'), (df_excel['관서명'] == '수서서') | (df_excel['관서명'] == '강남서'), (df_excel['관서명'] == '강서서'),
    (df_excel['관서명'] == '서초서') | (df_excel['관서명'] == '방배서'), (df_excel['관서명'] == '서부서') | (df_excel['관서명'] == '은평서'),
    (df_excel['관서명'] == '중부서') | (df_excel['관서명'] == '남대문서'), (df_excel['관서명'] == '종로서') | (df_excel['관서명'] == '혜화서'),
    (df_excel['관서명'] == '용산서'), (df_excel['관서명'] == '성북서') | (df_excel['관서명'] == '종암서'), (df_excel['관서명'] == '동대문서'),
    (df_excel['관서명'] == '마포서'), (df_excel['관서명'] == '영등포서'), (df_excel['관서명'] == '성동서'), (df_excel['관서명'] == '동작서'),
    (df_excel['관서명'] == '광진서'), (df_excel['관서명'] == '강북서'), (df_excel['관서명'] == '금천서'), (df_excel['관서명'] == '중랑서'),
    (df_excel['관서명'] == '관악서'), (df_excel['관서명'] == '강동서'), (df_excel['관서명'] == '구로서'), (df_excel['관서명'] == '양천서'),
    (df_excel['관서명'] == '송파서'), (df_excel['관서명'] == '노원서'), (df_excel['관서명'] == '도봉서')]  # 관할서 리스트
gu_list = ['서대문구', '강남구', '강서구', '서초구', '은평구', '중구', '종로구', '용산구', '성북구', '동대문구', '마포구', '영등포구', '성동구',
           '동작구', '광진구', '강북구', '금천구', '중랑구', '관악구', '강동구', '구로구', '양천구', '송파구', '노원구', '도봉구']  # 구 리스트
df_excel['구별'] = np.select(ps_list, gu_list, default='구 없음')  # 관할서마다 순서에 맞는 구 볼러오기 (해당되는 게 없을 땐 '구 없음' 지정)
print(df_excel)  # 결과물 출력
```

```python
# 3-3 필수과제
# 구별을 인덱스로 한 피벗 테이블 생성
pivot = pd.pivot_table(df_excel, index='구별', values=['강간(검거)', '강간(발생)', '강도(검거)', '강도(발생)', '살인(검거)', '살인(발생)',
                                                    '소계(검거)', '소계(발생)', '절도(검거)', '절도(발생)', '폭력(검거)', '폭력(발생)'], aggfunc='sum')
print(pivot)  # 결과물 출력
```

```python
# 3-4 필수과제
df_excel = df_excel.drop(index=0, axis=0)  # 구 없음 행 삭제(첫번째 행)
print(df_excel)                            # 결과물 출력
```

```python
# 3-5 필수과제
pivot['강간검거율'] = pivot['강간(검거)'] / pivot['강간(발생)'] * 100  # 강간검거율 열 생성
pivot['강도검거율'] = pivot['강도(검거)'] / pivot['강도(발생)'] * 100  # 강도검거율 열 생성
pivot['살인검거율'] = pivot['살인(검거)'] / pivot['살인(발생)'] * 100  # 살인검거율 열 생성
pivot['절도검거율'] = pivot['절도(검거)'] / pivot['절도(발생)'] * 100  # 절도검거율 열 생성
pivot['폭력검거율'] = pivot['폭력(검거)'] / pivot['폭력(발생)'] * 100  # 폭력검거율 열 생성
pivot['검거율'] = pivot['소계(검거)'] / pivot['소계(발생)'] * 100      # 검거율 열 생성
print(pivot)                                                          # 결과물 출력
```

```python
# 3-6 필수과제
del pivot['강간(검거)']  # 강간(검거) 행 삭제
del pivot['강도(검거)']  # 강도(검거) 행 삭제
del pivot['살인(검거)']  # 살인(검거) 행 삭제
del pivot['절도(검거)']  # 절도(검거) 행 삭제
del pivot['폭력(검거)']  # 폭력(검거) 행 삭제
del pivot['소계(발생)']  # 소계(발생) 행 삭제
del pivot['소계(검거)']  # 소계(검거) 행 삭제
print(pivot)             # 결과물 출력
```

```python
# 3-7 필수과제
pivot.rename(columns = {'강간(발생)' : '강간', '강도(발생)' : '강도', '살인(발생)' : '살인', '절도(발생)' :' 절도',
                        '폭력(발생)' : '폭력'}, inplace = True)  # 컬럼 명 변경
print(pivot)                                                    # 결과물 출력
```

```python
# 3-1 선택과제
df_csv = pd.read_csv('pop_kor.csv')              # 인구 데이터 csv 불러옴
df_csv = df_csv.set_index('구별')                # 구별로 index 세팅
merged_df = pd.merge(df_csv, pivot, on = '구별') # merge 메서드로 구별을 기준으로 두 데이터프레임 합치기
print(merged_df)                                 # 결과물 출력
```

```python
# 3-2 선택과제
joined_df = df_csv.join(pivot) # join 메서드를 사용하여 두 데이터프레임 합치기
print(joined_df)               # 결과물 출력
```

```python
# 3-3 선택과제
sorted_df = joined_df.sort_values(by = '검거율') # 검거율 기준으로 오름차순 정렬 (by 뒤에 ascending=True 생략되어 있음)
print(sorted_df)                                 # 결과물 출력
```
